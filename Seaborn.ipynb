```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Flights Dataset

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
flights.isna().sum().sum() #Checking for any null values
```

```python
sns.relplot(x='year', y='passengers', hue='month', data=flights);
```

As the year increases we see that on average there's an increasing linear trend where the number of passengers per flight also increased. Moreover, you can see that on average for each year there were more passengers during the warmer months (i.e. May through September).

```python
flights_2 = flights.drop('month', axis = 1)
```

```python
sns.heatmap(flights_2)
```

```python
sns.catplot(x='year', y = 'passengers', kind='boxen', data = flights_2);
```
